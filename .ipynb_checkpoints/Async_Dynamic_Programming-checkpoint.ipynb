```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.integrate as integrate
```

```python
t = np.linspace(0,.5, 500)

f=20
orig_func = np.sin(2*np.pi*f*t)

# Calculating the distance along the sinusoidal based on this link
# https://www.quora.com/How-long-is-a-sine-wave#:~:text=The%20length%20of%20a%20sine,of%20150%20meters%20or%20more
def dist_map(start,end):
    return integrate.quad(lambda x: np.sqrt(1+(2*np.pi*f*np.cos(2*np.pi*f*x))**2),
      start, end)[0]
        
```

```python
N = 300
M = 200 
```

```python
plt.figure(figsize=(4,4))
plt.plot(orig_func)
```

```python
# Initializations
phi = np.zeros((M, N,N))
eta = np.zeros((M,N))
i=0
for n in range(N):    
    phi[0,i, n] = dist_map(i, n)
    eta[0,n] = i
    



```

```python
for n in range(N):
    for m in range(M-2):
        phi[m+1,i, n]=np.min(phi[m, i, :]+ [dist_map[i,l] for l in range(N)])
        
```
